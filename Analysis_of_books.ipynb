```python
books_cleaned=books[books['original_publication_year'].isnull()==False]
```

```python
books_cleaned.loc[:, 'original_publication_year'] = books_cleaned['original_publication_year'].astype(int)
```

```python
books_cleaned=books_cleaned.drop(['goodreads_book_id','best_book_id','original_title','work_id','isbn','isbn13','image_url','small_image_url'],axis=1)
```

```python
books_cleaned=books_cleaned[books_cleaned['language_code'].isnull()==False]
```

```python
books_cleaned.reset_index(drop=True, inplace=True)
```

```python
books_cleaned['language_code'] = books_cleaned['language_code'].replace(['en-US', 'en-CA', 'en-GB'], 'eng')
```

```python
# Each rating multiplied by its star count, summed up, then divided by the total number of ratings
books_cleaned['average_rating'] = (
    (1 * books_cleaned['ratings_1'] +
     2 * books_cleaned['ratings_2'] +
     3 * books_cleaned['ratings_3'] +
     4 * books_cleaned['ratings_4'] +
     5 * books_cleaned['ratings_5']) /
    (books_cleaned['ratings_1'] +
     books_cleaned['ratings_2'] +
     books_cleaned['ratings_3'] +
     books_cleaned['ratings_4'] +
     books_cleaned['ratings_5'])
)
```

#Highest rated Authors in last 5 years

```python
import pandas as pd

last_year = books_cleaned['original_publication_year'].max()

# Filtering out the data for the last 5 years
recent_books = books_cleaned[books_cleaned['original_publication_year'] > last_year - 5]

# calculate the mean average_rating, grouped by year and author
author_ratings = recent_books.groupby(['original_publication_year', 'authors'])['average_rating'].mean()

author_ratings = author_ratings.reset_index()

# Finding the highest-rated author for each year
highest_rated_authors = author_ratings.loc[author_ratings.groupby('original_publication_year')['average_rating'].idxmax()]

print(highest_rated_authors)
```

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(10, 6))  # Sets the figure size
scatter = plt.scatter(highest_rated_authors['authors'], highest_rated_authors['average_rating'],
                      c=highest_rated_authors['original_publication_year'], cmap='viridis', s=100)  # s is the size of each point

# Adding labels and title
plt.title('Highest Rated Authors in the Last 5 Years')
plt.xlabel('Authors')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')  # Rotate labels to prevent overlap

# Adding a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Publication Year')

# Annotate each point with the corresponding year
for i, row in highest_rated_authors.iterrows():
    plt.annotate(row['original_publication_year'],
                 (row['authors'], row['average_rating']),
                 textcoords="offset points",  # how to position the text
                 xytext=(0,10),  # distance from text to points (x,y)
                 ha='center')  # horizontal alignment can be left, right or center

# Show the plot
plt.show()
```

#Highest rated books in last 5 years

```python
import pandas as pd

last_year = books_cleaned['original_publication_year'].max()

# Filtering out the data for the last 5 years
recent_books = books_cleaned[books_cleaned['original_publication_year'] > last_year - 5]

# calculate the mean average_rating, grouped by year and author
books_ratings = recent_books.groupby(['original_publication_year', 'title'])['average_rating'].mean()

books_ratings = books_ratings.reset_index()

# Finding the highest-rated author for each year
highest_rated_books = books_ratings.loc[books_ratings.groupby('original_publication_year')['average_rating'].idxmax()]

print(highest_rated_books)
```
