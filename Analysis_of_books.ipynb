```python
books_cleaned=books[books['original_publication_year'].isnull()==False]
```

```python
books_cleaned.loc[:, 'original_publication_year'] = books_cleaned['original_publication_year'].astype(int)
```

```python
books_cleaned=books_cleaned.drop(['goodreads_book_id','best_book_id','original_title','work_id','isbn','isbn13','image_url','small_image_url'],axis=1)
```

```python
books_cleaned=books_cleaned[books_cleaned['language_code'].isnull()==False]
```

```python
books_cleaned.reset_index(drop=True, inplace=True)
```

```python
books_cleaned.info()
```

```python
summary=books_cleaned.describe(include='all')
print(summary)
```

```python
books_cleaned['language_code'] = books_cleaned['language_code'].replace(['en-US', 'en-CA', 'en-GB'], 'eng')
```

```python
# Each rating multiplied by its star count, summed up, then divided by the total number of ratings
books_cleaned['average_rating'] = (
    (1 * books_cleaned['ratings_1'] +
     2 * books_cleaned['ratings_2'] +
     3 * books_cleaned['ratings_3'] +
     4 * books_cleaned['ratings_4'] +
     5 * books_cleaned['ratings_5']) /
    (books_cleaned['ratings_1'] +
     books_cleaned['ratings_2'] +
     books_cleaned['ratings_3'] +
     books_cleaned['ratings_4'] +
     books_cleaned['ratings_5'])
)
```

#Highest rated Authors in last 5 years

```python
import pandas as pd

last_year = books_cleaned['original_publication_year'].max()

# Filtering out the data for the last 5 years
recent_books = books_cleaned[books_cleaned['original_publication_year'] > last_year - 5]

# calculate the mean average_rating, grouped by year and author
author_ratings = recent_books.groupby(['original_publication_year', 'authors'])['average_rating'].mean()

author_ratings = author_ratings.reset_index()

# Finding the highest-rated author for each year
highest_rated_authors = author_ratings.loc[author_ratings.groupby('original_publication_year')['average_rating'].idxmax()]

print(highest_rated_authors)
```

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(10, 6))  # Sets the figure size
scatter = plt.scatter(highest_rated_authors['authors'], highest_rated_authors['average_rating'],
                      c=highest_rated_authors['original_publication_year'], cmap='viridis', s=100)  # s is the size of each point

# Adding labels and title
plt.title('Highest Rated Authors in the Last 5 Years')
plt.xlabel('Authors')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')  # Rotate labels to prevent overlap

# Adding a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Publication Year')

# Annotate each point with the corresponding year
for i, row in highest_rated_authors.iterrows():
    plt.annotate(row['original_publication_year'],
                 (row['authors'], row['average_rating']),
                 textcoords="offset points",  # how to position the text
                 xytext=(0,10),  # distance from text to points (x,y)
                 ha='center')  # horizontal alignment can be left, right or center

# Show the plot
plt.show()
```

2. Gradual Increase:
  * Starting around the mid-20th century, there is a gradual increase in the number of books published annually. This trend continues, with more noticeable growth starting from the late 1970s.

3. Rapid Growth in Recent Decades:

  * The most striking feature of the graph is the sharp rise in the number of books published annually starting in the late 1990s, peaking dramatically around 2020. The numbers grow exponentially in these final decades, suggesting a significant shift in publishing practices or market dynamics.














# Annual Comparison of Book Publications by Language

```python
# Group by year and language, then count the number of books
language_trends = books_cleaned.groupby(['original_publication_year', 'language_code']).size().reset_index(name='count')
```

```python
language_trends
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for major languages, for example, English, Spanish, French
major_languages = language_trends[language_trends['language_code'].isin(['eng', 'spa', 'ger','ind'])]

# Create a line plot
plt.figure(figsize=(14, 7))
sns.lineplot(data=major_languages, x='original_publication_year', y='count', hue='language_code', marker='o')
plt.title('Publishing Trends by Language Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Books Published')
plt.legend(title='Language Code')
plt.show()
```

# Summary:

* The graph clearly highlights the dominance of the English language in publishing as compared to Spanish, Indonesian, and German.
* The exponential rise in publications in English near the 2020s suggests significant changes in publishing technology or market dynamics during this period.
* The presence of other languages like Spanish and German is much less pronounced, showing no significant trends or changes over the years.
* Indonesian publications are notably scarce, suggesting it may be underrepresented in the data.

# Number of books published by each authors and their average rating

```python
# Group by 'authors' and aggregate both the count of books and the average of the 'average_rating'
books_authors = books_cleaned.groupby('authors').agg({
    'book_id': 'size','average_rating': 'mean'}).reset_index().rename(columns={'book_id': 'count', 'average_rating': 'average_rating'})

books_authors = books_authors.sort_values(by='count', ascending=False)
books_authors.head(15)
```

```python
import plotly.express as px

fig = px.scatter(books_authors, x='count', y='average_rating',
                 hover_data=['authors'], title='Interactive Plot of Book Count and Average Ratings')
fig.show()
```

# The key observations are:

* Most authors have published between 1 and 5 books, with fewer authors having higher counts.
* Average ratings are mostly clustered between 3.8 and 4.2, indicating a generally favorable reception of their works.
* Vertical bands at specific book counts suggest that many authors share common publication outputs (e.g., 1, 2, 3 books).
* There is no clear trend that higher book counts correlate with lower or higher average ratings; authors with both low and high book counts can maintain high average ratings.

Overall, the plot shows a wide distribution of book counts with relatively stable average ratings across different levels of author productivity.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot for text reviews vs. average rating
plt.figure(figsize=(8, 4))
sns.regplot(x='work_text_reviews_count', y='average_rating', data=books_cleaned, scatter_kws={'alpha':0.5})
plt.title('Correlation between Text Reviews Count and Average Rating')
plt.xlabel('Number of Text Reviews')
plt.ylabel('Average Rating')
plt.show()

# Plot for text reviews vs. total ratings count
plt.figure(figsize=(8, 4))
sns.regplot(x='work_text_reviews_count', y='ratings_count', data=books_cleaned, scatter_kws={'alpha':0.5})
plt.title('Correlation between Text Reviews Count and Total Ratings Count')
plt.xlabel('Number of Text Reviews')
plt.ylabel('Total Ratings Count')
plt.show()
```

#Top Plot: Correlation between Text Reviews Count and Average Rating


* There is a slight positive correlation between the number of text reviews and the average rating of books. As the number of text reviews increases, the average rating tends to increase slightly.
* The data points are densely clustered at the lower end of the text reviews count, indicating that most books receive few reviews. However, books with more reviews tend to have higher average ratings, though the increase is gradual.
* There are fewer books with very high numbers of reviews, but these tend to have relatively high average ratings.

#Bottom Plot: Correlation between Text Reviews Count and Total Ratings Count


* This plot shows a much stronger positive correlation compared to the first plot. As the number of text reviews increases, the total ratings count also increases substantially.
* The relationship appears more linear, with a clear upward trajectory as the number of text reviews grows.
* Books with a large number of text reviews also have a high total ratings count, suggesting that books which engage readers to leave text reviews also attract a larger number of ratings overall.
